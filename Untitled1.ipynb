```python
# ===============================================
# AI CREATIVE STUDIO - COMPLETE CODE
# ===============================================

# 1. Mount Google Drive (save your work)
from google.colab import drive
drive.mount('/content/drive')
print("✅ Google Drive mounted!")

# 2. Install libraries
# Ensure CUDA-enabled PyTorch is installed by uninstalling any existing version and reinstalling.
!pip uninstall -y torch torchvision torchaudio
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
!pip install diffusers transformers accelerate gradio pillow numpy requests safetensors
print("✅ Libraries installed!")

# Verify CUDA availability
import torch
if not torch.cuda.is_available():
    raise AssertionError("CUDA is not available. Please ensure your Colab runtime is set to GPU (Runtime -> Change runtime type -> Hardware accelerator: GPU).")
else:
    print(f"✅ CUDA is available! Device: {torch.cuda.get_device_name(0)}")

# 3. Import libraries
from diffusers import StableDiffusionPipeline
from PIL import Image
import time

print("Loading AI model... this may take 1-2 minutes.")
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    use_safetensors=True
)
pipe = pipe.to("cuda")
print("✅ Model loaded! Ready to create art.")

# 4. Generate image function
def generate_image(prompt="cyberpunk city with neon lights", steps=30, guidance=7.5, seed=42):
    generator = torch.Generator("cuda").manual_seed(seed)
    with torch.no_grad():
        image = pipe(
            prompt,
            num_inference_steps=steps,
            guidance_scale=guidance,
            generator=generator
        ).images[0]
    return image # Corrected: return the PIL Image directly

# 5. Run it!
print("\n" + "="*50)
print("🎨 GENERATING YOUR FIRST AI IMAGE...")
print("="*50)
image = generate_image()
print("✅ Image generated!")

# 6. Save and display
image.save("/content/drive/MyDrive/ai-creative-studio/first_image.png")
print("✅ Image saved to Google Drive!")
display(image)
print("\n🎉 SUCCESS! You just made AI art! ✨")
```
